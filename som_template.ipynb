```python
neurons=m.neurons
neurons
```

```python
def distance(a,b):
    res=0
    for i in range(len(a)):
        res += (a[i]-b[i])**2
    return res**0.5
```

```python
def best_matching_neuron(row,neurons):
    ans=[0,0]
    dis=float("inf")
    for i in range(len(neurons)):
        cord = m.coordinate(i)
        dist= distance(row,neurons[i])
        print("dis",dis,"ans",ans)
        if dist<dis:
            dis=dist
            ans[0]=cord[0]
            ans[1]=cord[1]
        
    return ans
    
        
```

```python
def class_pred(bmn,reference_nodes):
    arr=[distance(bmn,reference_nodes[i]) for i in range(3)]
    return arr.index(min(arr))
```
